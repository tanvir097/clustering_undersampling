```python
import os
import numpy as np
from sklearn.metrics import roc_curve, precision_recall_curve, f1_score, matthews_corrcoef
from scipy import interp
import pandas as pd
import seaborn as sns
# from sklearn.ensemble import AdaBoostClassifier

# from CUSBoost import CUSBoostClassifier
# from rusboost import RusBoost
from cus_sampling import cus_sampler
from cus_sampling_new import cus_sampler_new
from cus_sampling_new_2 import cus_sampler_new_2
from cus_sampling_new_3 import cus_sampler_new_3
from cus_sampling_new_6 import cus_sampler_new_6
from cus_sampling_new_7 import cus_sampler_new_7
from cus_sampling_new_8 import cus_sampler_new_8
from cus_sampling_new_9 import cus_sampler_new_9

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import Normalizer,MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import math
from sklearn.svm import SVC
from sklearn.metrics import auc
from sklearn.ensemble import AdaBoostClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import label_binarize

from sklearn.metrics import roc_auc_score, average_precision_score
from sklearn.metrics import matthews_corrcoef, f1_score, accuracy_score

from sklearn.model_selection import StratifiedKFold, KFold
```

```python
classifier = AdaBoostClassifier(n_estimators = 65, learning_rate= 0.1)
cluster_number = [5,10,15,20,23,25,30]
labelencoder = LabelEncoder()
normalization_object = Normalizer()
```

```python
result = pd.read_excel('result.xlsx')
result
```

```python
for file in file_list:
    model_dir = 'new_data_2/' + file
    df = pd.read_csv(model_dir, header = None)
    
    df['class'] = df[df.shape[1] - 1]
    df.drop([df.shape[1] - 2], axis=1, inplace=True)
    
    df['class'] = labelencoder.fit_transform(df['class'])
    if file == 'abalone9-18.txt' or file == 'abalone19.txt':
        df[0] = labelencoder.fit_transform(df[0])
        
    X = np.array(df.drop(['class'], axis=1))
    y = np.array(df['class'])
    
    X = normalization_object.fit_transform(X)
    skf = StratifiedKFold(n_splits=5, shuffle=True)
    
    
    for cluster in cluster_number:
        acc = []
        roc_auc = []
        auc_pr = []
        f1 = []
        mcc = []
        data = []

        for train_index, test_index in skf.split(X, y):
            X_train = X[train_index]
            X_test = X[test_index]

            y_train = y[train_index]
            y_test = y[test_index]

            X_train_sample, y_train_sample, _ = cus_sampler(X_train,y_train, number_of_clusters = cluster)
    #         X_train_sample, y_train_sample, _ = cus_sampler_new_5(X_train,y_train)

        #     print(len(y_train_sample[y_train_sample==0]))
        #     print(len(y_train_sample[y_train_sample==1]))
            classifier.fit(X_train_sample, y_train_sample)

        #     y_bin_test = label_binarize(y_test, classes=n_classes)
            y_pred = classifier.predict(X_test)
            y_pred_proba =classifier.predict_proba(X_test)
            precision, recall, thresholds = precision_recall_curve(y_test, y_pred_proba[:, 1])

            acc.append(accuracy_score(y_test, y_pred))
            f1.append(f1_score(y_test, y_pred, average='macro'))
            mcc.append(matthews_corrcoef(y_test, y_pred))
            roc_auc.append(roc_auc_score(y_test, y_pred_proba[:, 1])) # [:, 1] multi_class='ovr'
            auc_pr.append(auc(recall, precision))
        #     auc_pr.append(aucpr(y_bin_test, y_pred_proba))

        Accuracy= round((np.mean(acc)),4)
        F1_Score= round((np.mean(f1)),4)
        MCC= round((np.mean(mcc)),4)
        Auc_PR= round((np.mean(auc_pr)),4)
        Roc_AUC= round((np.mean(roc_auc)),4)
        
        data.append(file)
        data.append('baseline')
        data.append(cluster)
        data.append(Accuracy)
        data.append(F1_Score)
        data.append(MCC)
        data.append(Auc_PR)
        data.append(Roc_AUC)

#         print(data)

        result.loc[len(result.index)] = data
```
